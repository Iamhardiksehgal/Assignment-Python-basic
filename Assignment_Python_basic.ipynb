# Assignment : Python basic

1. What is Python, and why is it popular ?
   - Python is a high-level, interpreted programming language known for its simple and readable syntax. It is popular because it is easy to learn, has a large standard library, supports multiple programming paradigms, and has a strong community. Python is widely used in web development, data science, artificial intelligence, and automation due to its versatility and extensive third-party libraries.

2. What is an interpreter in Python ?
   - An interpreter in Python is a program that reads and executes Python code line by line. It translates the high-level Python code into machine code at runtime, allowing the program to run without being compiled first. This makes Python easier to test and debug.

3. What are pre-defined keywords in Python ?
   - Pre-defined keywords in Python are reserved words that have special meaning in the language. They are used to define the syntax and structure of Python programs. These keywords cannot be used as variable names or identifiers because they are already assigned specific functions by the Python interpreter.

4. Can keywords be used as variable names ?
   - No, keywords cannot be used as variable names in Python. Keywords are reserved words in Python that have special meaning and are used to define the syntax and structure of the language. Using them as variable names will cause a syntax error.

5. What is mutability in Python ?
   - Mutability in Python refers to the ability of an object to be changed or modified after it is created. Based on this property, Python objects are categorized into mutable and immutable types.

6. Why are lists mutable, but tuples are immutable ?
   - Lists are mutable because they are designed to hold a collection of items that may need to be changed, such as adding, removing, or updating elements. This flexibility makes them useful for dynamic data. On the other hand, tuples are immutable to provide stability and efficiency, especially when storing fixed data that should not be changed. Their immutability allows Python to optimize memory usage and makes tuples faster and safer to use in certain situations like dictionary keys.

7. What is the difference between “==” and “is” operators in Python ?
   - In Python, == checks if the values of two variables are equal, while is checks if two variables point to the same object in memory. For example, if a = [1, 2] and b = [1, 2], then a == b will return True because the values are the same, but a is b will return False because they are two different objects in memory. So, == compares content, whereas is compares identity.        

8. What are logical operators in Python ?
   - Logical operators in Python are used to perform logical operations and combine conditional statements. The three main logical operators are: and, or, and not. The and operator returns True if both conditions are true, or returns True if at least one condition is true, and not reverses the result of the condition. These operators are commonly used in if statements and control flow to make decisions based on multiple conditions.

9. What is type casting in Python ?
   - Type casting in Python is the process of converting one data type into another, such as converting a string to an integer or a float to a string. This is useful when you need to perform operations between different data types. Python provides built-in functions like int(), float(), str(), and bool() to perform type casting. For example, int("5") will convert the string "5" into the integer 5.

10. What is the difference between implicit and explicit type casting ?
    - In Python, implicit type casting happens automatically when Python converts one data type to another during an operation, without user intervention—for example, converting an integer to a float during arithmetic (5 + 2.0 becomes 7.0). Explicit type casting is done manually by the programmer using functions like int(), float(), or str() to convert data types as needed—e.g., int("10") converts the string "10" to the integer 10.      
         
11. What is the purpose of conditional statements in Python ?
    - The purpose of conditional statements in Python is to control the flow of the program by allowing it to make decisions based on certain conditions. They let the program execute specific blocks of code only when a condition is true, using statements like if, elif, and else. This helps in creating dynamic and responsive programs that behave differently based on different inputs or situations.

12. How does the elif statement work ?
    - The elif (short for "else if") statement in Python is used to check multiple conditions one after another after an initial if statement. If the if condition is false, Python checks the elif condition; if it’s true, that block runs and the rest are skipped. You can have multiple elif statements to handle different cases. It works like a chain: Python evaluates each condition in order and stops as soon as one is true. If none are true, the optional else block (if used) will execute.             

13. What is the difference between for and while loops ?
    - In Python, a for loop is used when you want to iterate over a sequence like a list, tuple, string, or range for a known number of times, while a while loop is used when you want to repeat a block of code as long as a condition is true, and you may not know beforehand how many times it will run. The for loop is ideal for fixed iterations, and the while loop is better suited for condition-based repetition.

14. Describe a scenario where a while loop is more suitable than a for loop ?
    - A while loop is more suitable than a for loop when the number of iterations is unknown and depends on a condition. For example, if you're building a login system that keeps asking the user to enter the correct password until they get it right, a while loop is ideal because you don’t know how many attempts the user will need. It keeps running as long as the condition (wrong password) is true, making it perfect for such situations.
